```python
import pandas as pd

df = pd.read_csv("orders.csv")
```

```python
data = df.head(10)
```

```python
data.sort_index(ascending=False) #sort increseing order by using index
```

```python
data.sort_values("sales")  # icreseing order
```

```python
data.sort_values("sales", ascending= False) #decresing order
```

```python
# sort by columns
data.sort_values("category")
```

```python
# sort by multiple columns
data.sort_values(["category", "sales"])
```

```python
data.sort_values(["category", "sales"], ascending= [True, False])
```

```python
# Filtering data

data.loc[data["region"] == "West"]
```

```python
data.loc[data["region"] == "South", ["order_id", "order_date"]] # sort selected columns using ccondition
```

```python
data.loc[(data["region"] == "South") & (data["profit"] > 0)] # sort with multiple condition
```

```python
data[data["region"] == "South"] # sort withoout using loc
```

```python
~(data["profit"] > 0)   # inverse the condition
```
